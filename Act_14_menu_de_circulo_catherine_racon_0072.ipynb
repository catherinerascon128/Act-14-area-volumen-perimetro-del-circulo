 1. ▶ Área del Círculo:      A = π * r^2
 2. ▶ Perímetro del Círculo: P = 2 * π * r
 3. ▶ Volumen de la Esfera:  V = (4/3) * π * r^3

* ☣ Donde r representa el radio de la figura y r (pi) es la constante matemática aproximadamente igual a 3.1415926535.. desarrollado un en Python estructurado de forma modular programa.
* ✒ Para implementar estos cálculos de manera práctica, se ha
* ⏰ EIl programa utiliza un menú interactivo que se mantiene en ejecución mediante un ciclo while, permitiendo al usuario realizar múltiples calculos sin reiniciar la aplicacion.
* ♒ La navegación se controla con estructuras condicionales if, elif y else, las cuales dirigen el flujo hacia la funcion correspondiente según la opción elegida
Cada operación está encapsulada en su propia función (areacirculo(), perimetrocirculo( y volumenesfera(), cumpliendo con el principlo de responsabilidad única y facilitando la lectura del código para estudiantes de nivel principiante.
* Al invocar cada función, se solicita irectamente el valor del radio, se aplica la formula matemética utilizando la libreria eståndar math y se muestra el resultado con un formato limpio y decorado con emojis
* Este enfoque no solo refuerza el pensamiento logico y la sintaxis básica de Python, sino que también demuestra cómo la programación puede servir como herramienta práctica para resolver problemas geométricos de forma rápida y precisa.

```python
# -----------------------------------------------------------------------------
# Fórmulas Matemáticas Utilizadas:
# 1. Área del Círculo:      A = π * r^2
# 2. Perímetro del Círculo: P = 2 * π * r
# 3. Volumen de la Esfera:  V = (4/3) * π * r^3
# -----------------------------------------------------------------------------

import math

def areacirculo():
    """Calcula y muestra el área de un círculo dado el radio."""
    print("\n--- 🟢 Cálculo del Área del Círculo ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: A = π * r^2
    area = math.pi * (radio ** 2)

    print(f"✨ El área del círculo con radio {radio} es: {area:.2f}")

def perimetrocirculo():
    """Calcula y muestra el perímetro (circunferencia) de un círculo dado el radio."""
    print("\n--- 📏 Cálculo del Perímetro del Círculo ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: P = 2 * π * r
    perimetro = 2 * math.pi * radio

    print(f"✨ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera dado el radio."""
    print("\n--- 🔵 Cálculo del Volumen de la Esfera ---")
    radio = float(input("Por favor, ingresa el radio de la esfera: "))

    # Fórmula: V = (4/3) * π * r^3
    volumen = (4/3) * math.pi * (radio ** 3)

    print(f"✨ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

def menu():
    """Muestra las opciones disponibles al usuario."""
    print("\n" + "="*40)
    print("📐 MENÚ DE GEOMETRÍA 📐")
    print("="*40)
    print("1️⃣  Calcular Área del Círculo")
    print("2️⃣  Calcular Perímetro del Círculo")
    print("3️⃣  Calcular Volumen de la Esfera")
    print("4️⃣  Salir del programa")
    print("="*40)

# Programa Principal
def main():
    opcion = 0

    # Bucle while para mantener el programa activo hasta que el usuario quiera salir
    while opcion != 4:
        menu()  # Llamada a la función menú

        try:
            # Se pide la opción al usuario
            opcion = int(input("👉 Selecciona una opción (1-4): "))

            # Estructura condicional if, elif, else
            if opcion == 1:
                areacirculo()       # Llamada a función área
            elif opcion == 2:
                perimetrocirculo()  # Llamada a función perímetro
            elif opcion == 3:
                volumenesfera()     # Llamada a función volumen
            elif opcion == 4:
                print("\n👋 ¡Gracias por usar el programa! Hasta luego. 🚀\n")
            else:
                # Caso para números fuera del rango 1-4
                print("\n⚠️  Opción no válida. Por favor, selecciona un número del 1 al 4.")

        except ValueError:
            # Aunque no se pide validar exhaustivamente, esto evita que el programa
            # se cierre si el usuario escribe letras en el menú principal por error.
            print("\n⚠️  Por favor, ingresa un número entero válido para el menú.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
